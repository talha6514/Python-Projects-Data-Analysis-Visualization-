### Choropleth for USA

```python
import chart_studio.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import pandas as pd
```

```python
init_notebook_mode(connected=True)
```

```python
data = dict(type = 'choropleth',
            locations = ['AZ','CA','NY'],
            locationmode = 'USA-states',
            colorscale = 'Portland',
            text = ['text1','text2','text3'],
            z = [1.0,2.0,3.0],
            colorbar = {'title':'Colorbar Title Goes Here'})
```

```python
layout = dict(geo = {'scope':'usa'})
```

```python
choromap = go.Figure(data = [data], layout = layout)
```

```python
iplot(choromap)
```

### Choropleth for "2011_US_AGRI_Exports" (From Original Data)

```python
df1 = pd.read_csv('2011_US_AGRI_Exports')
```

```python
df1.head()
```

```python
data = dict(type = 'choropleth', 
            colorscale = 'ylorbr', 
            locations = df1['code'], 
            locationmode = 'USA-states',
            z = df1['total exports'],
            text = df1['text'],
            marker = dict(line = dict(color = 'rgb(255,255,255)', width = 2)),
            colorbar = {'title':'Million USD'})
```

```python
choromap2 = go.Figure(data = [data], layout = layout)
```

```python
iplot(choromap2)
```

### Choropleth Maps Part2 - International (World GDP Dataset) (Real Data)

```python
df2 = pd.read_csv('2014_World_GDP')
```

```python
df2.head()
```

```python
data = dict(type = 'choropleth',
            locations = df2['CODE'],
            z = df2['GDP (BILLIONS)'],
            text = df2['COUNTRY'],
            colorbar = {'title':'GDP in Billion USD'})
```

```python
layout1 = dict(title = '2014 Global GDP',
              geo = dict(showframe = False, projection = {'type':'mercator'}))
```

```python
layout2 = dict(title = '2014 Global GDP',
              geo = dict(showframe = True, projection = {'type':'mercator'}))
```

```python
choromap3 = go.Figure(data = [data], layout = layout1)
```

```python
iplot(choromap3)
```

```python
choromap4 = go.Figure(data = [data], layout = layout2)
```

```python
iplot(choromap4)
```
