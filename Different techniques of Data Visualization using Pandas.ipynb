### Histogram Plot

```python
# Histogram of all the values in col A
df1['A'].hist()
```

```python
df1['A'].hist(bins = 30)
```

```python
# General method of calling the Plots
df1['A'].plot(kind = 'hist', bins = 30)
```

```python
# We can also call them like this
df1['A'].plot.hist(bins = 30)
```

### Area Plot

```python
df2.plot.area()
```

```python
df2.plot.area(alpha = 0.5)
```

### Bar Plot

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked = True, alpha = 0.5)
```

### Hexagonal Plot

```python
# For this we will be needing a new data frame
df = pd.DataFrame(np.random.randn(1000,2), columns = ['a','b'])
```

```python
df
```

```python
df.head()
```

```python
df.plot.hexbin(x = 'a', y = 'b', gridsize = 12)
```

```python
df.plot.hexbin(x = 'a', y = 'b', gridsize = 12, cmap = 'coolwarm')
```
